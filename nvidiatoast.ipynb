```python
# Create widgets that allow users to select the amount of time for their toast
toasty = widgets.Text(value='toast')
timer = widgets.IntSlider(description='time:', min=1, max=4)
display(toasty, timer)
```

```python
print(toasty.value + ' is set to ' + str(timer.value) + ' minute(s)')
# record this value to be set as the timer
timer = time.time() + (timer.value * 60)
```

```python
while time.time() < timer:
  print(timer - time.time())
```

```python
# while timer is on going, play a video
while time.time() < timer:
 # Play video from a file
  cap = cv2.VideoCapture('/content/Bread.mp4') # Remember to change video path before testing
# audio extension
player = MediaPlayer('/content/Bread.mp4')

# Throw error if video is not opened
if(cap.isOpened() == False):
    print("Error opening video file")
    print("file exists? Reason:", os.path.exists('/content/Bread.mp4'))
    sys.exit()

# Get the frames per second of the video
while(cap.isOpened()):
    ret, frame = cap.read()
    audio_frame, val = player.get_frame()
    if ret == True:
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            if val != 'eof' and audio_frame is not None:
                #audio
                img, t = audio_frame
            break
    else:
        break

# Release the video capture object
cap.release()
```
